```python
import torch
from matplotlib import pyplot as plt
```

```python
# set output format to svg
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
weight = torch.linspace(-1, 1, 500)
hard_weight = torch.sign(torch.nn.functional.hardshrink(weight, lambd=0.4))
```

```python
cmap = plt.get_cmap('viridis')

for alpha in torch.linspace(0,1,6):
    alpha = float(alpha)
    blend = weight * (1-alpha) + hard_weight * alpha
    plt.plot(weight, blend, label=f'$\\alpha = {alpha:.2}$', color=cmap(1-alpha))

plt.xlabel('Real weight')
plt.ylabel('Effective weight')
plt.legend()
plt.show()
```
