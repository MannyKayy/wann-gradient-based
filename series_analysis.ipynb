```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# set output format to svg
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
df = pd.concat([pd.read_csv(fn) for fn in ('learning_rate_2/results.csv', 'learning_rate_3/results.csv')])
df['growing'] = df['layer_sizes'].isnull()
```

```python
sns.swarmplot(data=df, x='lr', y='accuracy', hue='update_mechanism')
plt.show()
```

```python
sns.swarmplot(data=df, x='lr', y='accuracy', hue='growing')
plt.show()
```

```python
for lr in df['lr'].unique():
    subdf = df[df['lr'] == lr]
    print (lr, subdf['accuracy'].mean())
```
