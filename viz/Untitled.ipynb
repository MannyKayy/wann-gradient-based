```python
from model import Model
import torch
import matplotlib.pyplot as plt

from model_viz import graph_from_model, draw_model_graph
```

```python
# set output format to svg
from IPython.display import set_matplotlib_formats
%matplotlib inline
set_matplotlib_formats('svg')
```

```python
d = torch.load('model.pt')
shared_weight = torch.autograd.Variable(torch.Tensor([1]))
model = Model.from_dict(shared_weight, d['state'])
```

```python
g = graph_from_model(model)
edges = draw_model_graph(g)
print(edges)
plt.gcf().set_size_inches(10,4)

plt.show()
```

```python
model.cleanup()
g = graph_from_model(model)
draw_model_graph(g)
plt.gcf().set_size_inches(10,4)
plt.show()
```
